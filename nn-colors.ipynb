Let's setup tensorflow

```python
input_vector = tf.placeholder(tf.float32, [None, 6])

def model(X, w_h, w_h2, w_o):
    h = tf.nn.relu(tf.matmul(X, w_h))
    h2 = tf.nn.relu(tf.matmul(h, w_h2))
    return tf.matmul(h2, w_o)

w_h = tf.Variable(tf.random_normal([6, 900], stddev=0.01))
w_h2 = tf.Variable(tf.random_normal([900, 900], stddev=0.01))
w_o = tf.Variable(tf.random_normal([900, 3], stddev=0.01))

output_vector = model(input_vector, w_h, w_h2, w_o)

correct_answer_vector = tf.placeholder(tf.float32, [None, 3])
```

```python
cost_function = tf.reduce_sum(tf.square(correct_answer_vector - output_vector))
train_step = tf.train.GradientDescentOptimizer(0.001).minimize(cost_function)
```

```python
sess = tf.InteractiveSession()
sess.run(tf.initialize_all_variables())
```

Prepare the training data

```python
with open("colors.json") as f:
    color_raw_data = json.load(f)
    color_data = []
    for color_match in color_raw_data:
        color_data.append([list(map(lambda x: 1.0 * x / 256, colors.hex_to_rgb(color_match[0]) + colors.hex_to_rgb(color_match[1]))),
                      list(map(lambda x: 1.0 * x / 256, colors.hex_to_rgb(color_match[2])))])
```

```python
print(color_raw_data[0])
print(color_data[0])
```

```python
def train(runs):
  costs = []
  for i in range(runs):
      batch_indices = random.sample(range(len(color_data)), 100)
      batch = list(map(lambda x: color_data[x], batch_indices))
      batch_xs = list(map(lambda x: x[0], batch))
      batch_ys = list(map(lambda x: x[1], batch))

      _, cost = sess.run([train_step, cost_function], feed_dict={input_vector: batch_xs, correct_answer_vector: batch_ys})
      costs.append(cost)
  plt.figure(1)
  plt.plot(range(runs), costs)
```

```python
def mix_colors(colorA, colorB):
    mixed = sess.run(output_vector, feed_dict={input_vector: [list(map(lambda x: 1.0 * x / 256, colors.hex_to_rgb(colorA) + colors.hex_to_rgb(colorB)))]})
    mixed_hex = colors.rgb_to_hex(list(map(lambda x: int(min(x,1) * 255), mixed[0])))
    return mixed_hex
```

```python
train(5)
```

```python
colors.draw_colors("#ce28d7", "#3db64d", mix_colors("#ce28d7", "#3db64d"))
colors.draw_colors("#ee0000", "#ffffff", mix_colors("#ee0000", "#ffffff"))
```

```python
train(50)
```
